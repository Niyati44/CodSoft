# Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
```

# Loading Data

```python
df = data[['v1', 'v2']]
```

```python
df.head()
```

```python
df = df[['v1', 'v2']]
df.columns = ['label', 'message']
```

```python
df = df.dropna(subset=['label', 'message'])
```

```python
# Drop rows with any null values in 'label' or 'message'
df.dropna(subset=['label', 'message'], inplace=True)

# Map 'label' to binary values and remove any rows that didn't map correctly
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

# Drop rows with NaNs in 'label' after mapping
df.dropna(subset=['label'], inplace=True)

# Ensure 'label' is of integer type
df['label'] = df['label'].astype(int)
```

```python
df.head()
```

# Training Model

```python
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2, random_state=42)
```

```python
print("y_train type:", y_train.dtype)
print("y_test type:", y_test.dtype)
```

```python
print("X_train type:", X_train.dtype)
print("x_test type:", X_test.dtype)
```

```python
y_test.value_counts()
```

Applying TF-IDF Vectorization to differentiate spam messages.

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

Function to print result

```python
def train_and_evaluate_model(model, model_name):
    model.fit(X_train_tfidf, y_train)
    y_pred = model.predict(X_test_tfidf)
    print(f"{model_name} Evaluation:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("-" * 60)
```

Logistic ReRegression Model

```python
clf = LogisticRegression()
train_and_evaluate_model(clf, "Logistic Regression")
```

Multinomial Naive Bayes Model

```python
clf_nb = MultinomialNB()
train_and_evaluate_model(clf_nb, "Naive Bayes")
```

SVM Model

```python
# Support Vector Machine
clf_svm = SVC(kernel='linear', probability=True)
train_and_evaluate_model(clf_svm, "Support Vector Machine")
```

After comparing all models SVM is more accurate to identify spam sms.
